```python
from sklearn.preprocessing import StandardScaler
x_numeric = x.drop(columns=['ID'])
scaler = StandardScaler()


x= scaler.fit_transform(x_numeric)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score


rf = RandomForestClassifier()
rf.fit(xtrain, ytrain)

# Make predictions and evaluate the model
y_pred = rf.predict(xtest)
accuracy = accuracy_score(ytest, y_pred)
print(f'Random Forest Accuracy: {accuracy:.2f}')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
model = Sequential()
model.add(Dense(16, input_dim=xtrain.shape[1], activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(xtrain, ytrain, epochs=20, batch_size=10, verbose=1)

# Make predictions and evaluate the model
y_pred = (model.predict(xtest) > 0.5).astype("int32")
accuracy = accuracy_score(ytest, y_pred)
print(f'ANN Accuracy: {accuracy:.2f}')
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from skimage.measure import label, regionprops
from skimage.morphology import closing, square





img_path = r"C:\Users\user\Downloads\cervitester (2)\cervitester\smear2005\New database pictures\carcinoma_in_situ  \149143370-149143378-003-d.bmp"
img = Image.open(img_path)






img_rgb = img.convert("RGB")
img_rgb_array = np.array(img_rgb)




nucleus_color = np.array([0, 0, 255])  
cytoplasm_color = np.array([0, 0, 128])   









# Calculate the distance from each pixel to the nucleus and cytoplasm colors
nucleus_mask = np.all(np.abs(img_rgb_array - nucleus_color) < 50, axis=-1)
cytoplasm_mask = np.all(np.abs(img_rgb_array - cytoplasm_color) < 50, axis=-1)









# Apply morphological closing to fill small holes
nucleus_mask = closing(nucleus_mask, square(3))
cytoplasm_mask = closing(cytoplasm_mask, square(3))




# Label the regions
nucleus_label = label(nucleus_mask)
cytoplasm_label = label(cytoplasm_mask)









# Plot the processed masks
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(nucleus_mask, cmap='gray')
ax[0].set_title('Nucleus Mask')
ax[0].axis('off')
ax[1].imshow(cytoplasm_mask, cmap='gray')
ax[1].set_title('Cytoplasm Mask')
ax[1].axis('off')
plt.show()
```

```python
import numpy as np
from skimage.measure import label, regionprops

# Assuming nucleus_mask and cytoplasm_mask are the masks you have
# Label the regions
nucleus_label = label(nucleus_mask)
cytoplasm_label = label(cytoplasm_mask)

# Get properties of the labeled regions
nucleus_regions = regionprops(nucleus_label)
cytoplasm_regions = regionprops(cytoplasm_label)

def calculate_properties(region, region_type):
    properties = {}
    properties[f"{region_type} Area"] = region.area
    properties[f"{region_type} Perimeter"] = region.perimeter
    properties[f"{region_type} Major Axis Length"] = region.major_axis_length
    properties[f"{region_type} Minor Axis Length"] = region.minor_axis_length
    
    # Calculate Elongation based on bounding box dimensions
    properties[f"{region_type} Elongation"] = region.minor_axis_length / region.major_axis_length
    
    # Calculate Roundness
    if region.major_axis_length > 0:  # Assuming major_axis_length is the longest diameter
        circle_area = (np.pi / 4) * (region.major_axis_length ** 2)
        roundness = region.area / circle_area
    else:
        roundness = np.nan  # Handle division by zero

    properties[f"{region_type} Roundness"] = roundness

    return properties

# Calculate properties for the largest regions
nucleus_properties = calculate_properties(max(nucleus_regions, key=lambda x: x.area), "Nucleus") if nucleus_regions else {}
cytoplasm_properties = calculate_properties(max(cytoplasm_regions, key=lambda x: x.area), "Cytoplasm") if cytoplasm_regions else {}

# Calculate the additional combined properties
if nucleus_properties and cytoplasm_properties:
    combined_properties = {
        "Nucleus/Cytoplasm Area Ratio": nucleus_properties["Nucleus Area"] / cytoplasm_properties["Cytoplasm Area"]
    }
    
    combined_properties.update(nucleus_properties)
    combined_properties.update(cytoplasm_properties)
else:
    combined_properties = {}

# Print the properties
for prop, value in combined_properties.items():
    print(f"{prop}: {value}")
```

```python

combined_feature_vector = list(combined_properties.values())

combined_feature_vector

```

```python
# Build the feature vector in the correct order
combined_feature_vector = [
    combined_properties["Nucleus Area"],         # Kerne_A
    combined_properties["Cytoplasm Area"],       # Cyto_A
    combined_properties["Nucleus/Cytoplasm Area Ratio"], # K/C
    combined_properties["Nucleus Minor Axis Length"],    # KerneShort
    combined_properties["Cytoplasm Minor Axis Length"],  # CytoShort
    combined_properties["Nucleus Major Axis Length"],    # KerneLong
    combined_properties["Cytoplasm Major Axis Length"],  # CytoLong
    combined_properties["Nucleus Elongation"],           # KerneElong
    combined_properties["Nucleus Roundness"],            # KerneRund
    combined_properties["Cytoplasm Roundness"],          # CytoRund
    combined_properties["Cytoplasm Elongation"],         # CytoElong
    combined_properties["Nucleus Perimeter"],            # KernePeri
    combined_properties["Cytoplasm Perimeter"],          # CytoPeri
]
```
